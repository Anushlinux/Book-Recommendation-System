```python
import pandas as pd
import numpy as np

books = pd.read_csv('books.csv')
ratings = pd.read_csv('ratings.csv')
users = pd.read_csv('users.csv')
```

```python
ratings_with_name = ratings.merge(books, on = "ISBN")
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
```

```python
avg_rating = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
 
```

```python
avg_rating['avg_ratings'] = avg_rating['avg_ratings'].round().astype(int)
avg_rating
```

```python
popular_df = num_rating_df.merge(avg_rating, on = "Book-Title")
popular_df
```

```python
basic_df = popular_df[popular_df["num_ratings"]>=250].sort_values('avg_ratings', ascending=False).reset_index()
basic_df
```

```python
best_df = basic_df.merge(books, on = "Book-Title").drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-S', 'num_ratings', 'avg_ratings']]
```

```python
best_df.head(50)
```

```python
#Collabrative filtering based now ->

x = ratings_with_name.groupby("User-ID").count()['Book-Rating']>250

padhe_likhe = x[x].index 
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(padhe_likhe)]
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books = y[y].index
```

```python
finals_rating = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
finals_rating.drop_duplicates()
```

```python
pt = finals_rating.pivot_table(index = "Book-Title", columns= 'User-ID', values= 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
sorted(list(enumerate(similarity_score[0])), key = lambda x:x[1], reverse=True)[1:6]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse=True)[1:20]
    
    data = []
    for i in similar_items:
        
        item = []
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-S"].values))
        
        data.append(item)
        
    return data    
```

```python
recommend('Zoya')
```
