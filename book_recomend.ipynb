```python
import pandas as pd
import numpy as np 

books = pd.read_csv("Books.csv")
users = pd.read_csv("Users.csv")
ratings = pd.read_csv("Ratings.csv")
```

```python
books.head()
```

```python
ratings_with_name = pd.merge(ratings, books, on='ISBN')
```

```python
new_rating = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index().rename(columns={'Book-Rating': 'totalRatingCount'})[['Book-Title', 'totalRatingCount']]
```

```python
new_rating.head()
```

```python
new_rating = ratings_with_name.groupby("Book-Title")['Book-Rating'].mean().reset_index()
new_rating.rename(columns = {'Book-Rating' : 'Average-Rating'}, inplace = True)
```

```python
new_rating = pd.merge(new_rating, new_rating, on = 'Book-Title')    
```
